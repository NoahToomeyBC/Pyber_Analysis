```python
%matplotlib inline
```

```python
#Import dependencies 

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create Axis Labels
plt.plot(x_axis, y_axis, label = "Boston")
plt.xlabel("Months")
plt.ylabel("Fare ($)")
#Create y axis limit
plt.ylim(0, 45)
# Create A title
plt.title("PyBer Fare by Month")
#create a legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Recreate chart in OOP
fig = plt.figure()
ax = fig.add_subplot()
#set labels and formatting
ax.plot (x_axis, y_axis, marker ="^", color ="green", linewidth=2, label = "Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
# Set y limit to 0 to 45
ax.set_ylim(0,45)
# Create grid and legend
ax.grid()
ax.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "k", label = "New York")
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.invert_yaxis()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.legend()

```

```python
# Make plot points larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# create and format scatter plot chart
plt.scatter(y_axis, x_axis, c = "red", label = "Chicago",s=[i*3 for i in y_axis] )
plt.xlabel ("Fare ($)")
plt.ylabel ("Date")
plt.gca().invert_yaxis()
plt.xlim(0,45)

plt.legend()
```

```python
# Create and format scatter plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", alpha =0.8,  s=[i*5 for i in y_axis], edgecolors='black', linewidth=2, label = "Boston")
#make january on top
ax.invert_yaxis()
#Set Labels
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
#Set limit on x axis
ax.set_xlim(0,50)
ax.legend()
```

```python
fig, ax = plt.subplots()
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, explode=explode_values, colors = colors, labels=x_axis, autopct='%.1f%%', startangle=100, counterclock =False, shadow='True')
plt.show()
```
