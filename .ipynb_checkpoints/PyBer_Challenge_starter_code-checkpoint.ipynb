# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#Create data frames for urban, rural, and suburban cities
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
```

```python
#  1. Get the total rides for each city type
#Get total rides for urban cities
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"].sum()

#Get total rides for rural cities
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"].sum()

#Get total rides for suburban cities
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"].sum()

#get total rides for all cities
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
```

```python
# 2. Get the total drivers for each city type
#total drivers for urban cities
urban_total_drivers =  city_data_df[city_data_df["type"] == "Urban"].sum()["driver_count"]


#total drivers for rural cities
rural_total_drivers =  city_data_df[city_data_df["type"] == "Rural"].sum()["driver_count"]

#total drivers for suburban cities
suburban_total_drivers =  city_data_df[city_data_df["type"] == "Suburban"].sum()["driver_count"]

#create a series of 
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
```

```python
#  3. Get the total amount of fares for each city type
urban_total_fare = urban_cities_df.sum()["fare"]
rural_total_fare = rural_cities_df.sum()["fare"]
suburban_total_fare = suburban_cities_df.sum()["fare"]
total_fare = pyber_data_df.sum()["fare"]
total_fare_summary = pyber_data_df.groupby(["type"]).sum()["fare"]
```

```python
#  4. Get the average fare per ride for each city type. 
urban_average_fare = urban_total_fare / urban_ride_count 
suburban_average_fare = suburban_total_fare / suburban_ride_count
rural_average_fare = rural_total_fare / rural_ride_count
average_fare = total_fare_summary / total_rides 
```

```python
# 5. Get the average fare per driver for each city type. 
urban_average_driver_fare = urban_total_fare / urban_total_drivers
suburban_average_driver_fare = suburban_total_fare / suburban_total_drivers
rural_average_driver_fare = rural_total_fare / rural_total_drivers
average_fare_driver = total_fare_summary / total_drivers
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
          {"Total Rides": total_rides, 
          "Total Drivers": total_drivers,
          "Total Fares" : total_fare_summary,
           "Average Fare per Ride": average_fare,
           "Average Fare per Driver": average_fare_driver})
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
date_type_df = pyber_data_df.groupby(["type", "date"]).sum()[['fare']]
date_type_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
date_type_summary = date_type_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
date_type_summary_pivot = date_type_summary.pivot(index = "date", columns="type", values="fare")
date_type_summary_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_pivot = date_type_summary_pivot.loc['1/1/2019':'4/29/2019']
new_pivot
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_pivot.index = pd.to_datetime(new_pivot.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_rides_df = new_pivot.resample('W').sum()
weekly_rides_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_rides_df = weekly_rides_df.plot(figsize = (30,10))
# Add a title 
weekly_rides_df.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
weekly_rides_df.set_xlabel("Month")
weekly_rides_df.set_ylabel("Fare($USD)")

plt.savefig('analysis/PyBer_fare_summary.png')
plt.show()
plt.tight_layout()
```
